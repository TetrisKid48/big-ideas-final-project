# Traffic Sign Count
The purpose of this metric is to get a general idea of how many times someone driving in a neighborhood will encounter signs that will directly impact their drive. There are many different types of signs listed in the data and since it would take a long time to individually filter out each one that doesn't directly affect driving, however it was easy to find the largest amounts of ones that don't. For example, there were many examples of wayfinding signage, which wouldn't affect someone's speed or path if it's predetermined. I also eliminated parking indicators since they wouldn't be used unless the path starts or ends in that neighborhood.

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

data = pd.read_csv("https://data.wprdc.org/datastore/dump/d078a6b5-83a3-4723-a3a9-5371cfe1cc0c")

sm1 = data[data["description"] != "Wayfinding Signage"]
sm2 = data[data["description"] != "No Parking Anytime"]
sm3 = data[data["description"] != "Pay Parking"]
sm4 = data[data["description"] != "1ST Monday Street Cleaning"]

newSigns = pd.concat([sm1,sm2,sm3,sm4])
noDupes = newSigns.drop_duplicates()


finalSigns = noDupes["neighborhood"].value_counts()
print(finalSigns)

finalSigns.tail(10).plot(kind='barh')
plt.title = 'Neighborhoods with Least Amounts of Signs'
plt.ylabel = 'Neighborhood' 
plt.xlabel = 'Number of Signs'
plt.show()

finalSigns.head(10).plot(kind='barh')
plt.title = 'Neighborhoods with Greatest Amounts of Signs'
plt.ylabel = 'Neighborhood' 
plt.xlabel = 'Number of Signs'
plt.show()


```
