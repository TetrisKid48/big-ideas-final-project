## Speed Humps - Elias
This dataset includes all speed bumps in Pittsburgh up to 2021.

```python
import pandas as pd

speedHumps = pd.read_csv("speedhumps.csv", index_col="hump_id")
```

```python
speedHumps['neighborhood'].unique()
```

```python
df = speedHumps['neighborhood'].value_counts()
df.plot.bar(ylim=[0,18])
```

This chart shows how many speed bumps are in each neighborhood.

#### Final Results

1st - Squirrel Hill South

2nd - Carrick

3rd (tied) - North Oakland, Polish Hill

5th (tied) - Homewood South, Brookline, South Side Slopes, Perry North, Upper Hill
