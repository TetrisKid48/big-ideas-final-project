## Our Metric
Our metric is "anti-drivability." Or, the best neighborhood in Pittsburgh is the one that inconveniences cars the most. 

## Our Datasets / Sub-Metrics
#### Alex - Traffic Signs
Alex used this dataset to count the total number of traffic signs in each neighborhood that you have to react to while driving. These slow down drivers as they have to read the signs to continue.
#### Rory - Parking Meters
Rory found which neighborhoods have the highest parking prices, making driving to these areas more expensive.
#### Eli - Speed Humps & Traffic Speeds
Eli first counts the number of speed bumps in each neighborhood, which limits drivers' speed. Second, he shows which neighborhoods have the most roads where drivers went slower than the speed limit, showing anti-drivability in action.

## The Best Neighborhood
We concluded that Squirrel Hill is the best neighborhood according to our metrics. It placed consistently high up in every metric, even when seperated into North and South.
