```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("PittsburghParkingMeters.csv")
df = df.drop([7,16],axis = 0)
df = df.drop(['_id','id','hours','shape_area','rate','maxhours','days','objectid','shape_length'],axis = 1)

sorted_df = df.sort_values(by=['rates','parkinglen'], ascending = False)

graph_df = sorted_df
sorted_df['rates'] = sorted_df['rates'].str.replace('$', '')
sorted_df['rates'] = sorted_df['rates'].astype(float, errors='ignore')

print(sorted_df)

sorted_df.plot(x='area', y='rates', kind='bar', figsize=(7, 4), title='Hourly Rates')
```

Points:

1st - Downtown

2nd - North Shore, Oakland, Strip District

5th - Bakery Square, Squirrel Hill, Lawrenceville
